Logic Explanation:
- First check if the number is 0.
- If not, check if it’s negative.
- If positive, use modulus % to determine even or odd.
- This ensures all possible cases are covered.




```python
def classify_number(num):
    if num == 0:
        return "Zero"
    elif num < 0:
        return "Negative"
    elif num > 0:
        if num % 2 == 0:
            return "Positive Even"
        else:
            return "Positive Odd"
```

Logic Explanation:
- Compare a with both b and c.
- If a is greater or equal, return a.
- Otherwise check b against the others.
- If neither a nor b is largest, then c must be the largest.
- This conditional structure ensures correctness by exhaustive comparison.


```python
def largest_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
```

Logic Explanation:
- Input validation is done first (negative age is invalid).
- Then conditions classify into "Minor" or "Adult".
- This shows how conditionals prevent invalid inputs and ensure meaningful output.


```python
def classify_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    elif age < 18:
        return "Minor"
    else:
        return "Adult"
```

```python
def process_list(numbers):
    count_even = 0
    for num in numbers:
        if num == 0:
            break  # stop processing
        elif num < 0:
            continue  # ignore negative numbers
        elif num % 2 == 0:
            count_even += 1
    return count_even
```
